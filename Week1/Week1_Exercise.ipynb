Import libraries and load the data.

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
import matplotlib.cbook
warnings.filterwarnings("ignore",category=matplotlib.cbook.mplDeprecation)
```

```python
# utility function to format data
def func(pct, allvalues):
    absolute = int(pct / 100. * np.sum(allvalues))
    return "{:.1f}%\n({:d})".format(pct, absolute)
```

```python
oadf = pd.read_csv('obama-approval-ratings.csv')
oadf_sorted = oadf.sort_values('Approve',ascending=False)
```

Now We have data loaded to dataframe, lets start with Plotting charts

# Bar Chart

```python
fig1 = plt.figure(figsize=(10,6))
plt.bar('Issue', 'Approve',data=oadf_sorted)
plt.xticks(rotation=90)
plt.xlabel("Issue", size=15)
plt.ylabel("Approval Ratings", size=15)
plt.title("Obama Approval Rating by Issue", size=18)
plt.savefig("Python_Bar_Chart_ObamaApproval.png")
plt.show()
```

# Stacked Bar Chart

```python
fig2 = plt.figure()
labels=oadf_sorted['Issue']
oadf_sorted.index = oadf_sorted[['Issue']]
oadf_sorted[['Approve','Disapprove','None']].plot(kind='bar', stacked=True, rot=90, alpha=0.8, legend=True, label=labels)
plt.xticks(rotation=90)
plt.xlabel("Issue", size=15)
plt.ylabel("Approval Ratings", size=15)
plt.title("Obama Approval Rating by Issue", size=18)
plt.savefig("Python_Stacked_Bar_Chart_ObamaApproval.png")
plt.show()
```

# Pie Chart

```python
labels = oadf_sorted['Issue']
explode = oadf_sorted[['Approve']]

issueslist = ["Economy"]
outval = oadf_sorted[['Issue','Approve','Disapprove','None']]
outval2 = outval[outval.Issue.isin(issueslist)]
outval4 = outval2.melt(id_vars=['Issue'], value_vars=['Approve','Disapprove','None'])
plt.pie(outval4['value'], autopct='%1.1f%%', labels= outval4['variable'])
plt.title("Obama Approval Ratings on Economic Issues", size=18)
plt.savefig("Python_PieChart_ObamaApproval.png")
plt.show()
```

# Donut Chart

```python
labels = oadf_sorted['Issue']
explode = oadf_sorted[['Approve']]
fig = plt.figure(figsize=(10, 6))
#plt.pie(explode, autopct = lambda pct: func(pct, explode), labels=labels)
plt.pie(outval4['value'], autopct='%1.1f%%', labels= outval4['variable'])
plt.title("Obama Approval Ratings on Economic Issues", size=18)
my_circle=plt.Circle( (0,0), 0.7, color='0.85')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.savefig("Python_DonutChart_ObamaApproval.png")
plt.show()
```

# Line Chart

```python

plt.plot('Issue', 'Approve',data=oadf_sorted, label='Approval')
plt.plot('Issue', 'Disapprove',data=oadf_sorted, label='Disapproval')
plt.plot('Issue', 'None',data=oadf_sorted, label='Neutral')
plt.xticks(rotation=90)
plt.xlabel("Issue", size=15)
plt.ylabel("Ratings %", size=15)
plt.title("Obama Approval/Disapproval Rating Trajectoryby Issue", size=18)
leg = plt.legend(loc='upper center')
plt.savefig("Python_Line_Chart_ObamaApproveDisapprove.png")
plt.show()
```
